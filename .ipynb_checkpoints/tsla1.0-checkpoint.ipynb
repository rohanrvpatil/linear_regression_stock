```python
import pandas as pd
import numpy as np

df=pd.read_csv('TSLA.csv')
pd.set_option('display.precision', 5)
df.head()
```

```python
#sorting according to date
df['Date'] = pd.to_datetime(df['Date'], format='%d-%m-%Y') #converting object datatype to date datatype
cleaned_df=df.sort_values(by=["Date"])

cleaned_df
```

```python
#removing duplicates
cleaned_df=cleaned_df.drop_duplicates(subset=['Date'], ignore_index=True) #ignore_index puts index after removing duplicates

cleaned_df
```

```python
cleaned_df.info()
```

```python
pd.set_option('display.float_format', lambda x: '%.2f' % x)
cleaned_df.describe()
```

```python
#Check the categorical and numerical columns

# Categorical columns
cat_col = [col for col in cleaned_df.columns if cleaned_df[col].dtype == 'object']
print('Categorical columns :',cat_col)
# Numerical columns
num_col = [col for col in cleaned_df.columns if cleaned_df[col].dtype != 'object']
print('Numerical columns :',num_col)
```

```python
#Check the total number of unique values in the Categorical columns
cleaned_df[cat_col].nunique() #because finding unique numbers in numerical columns doesn't make sense
#cleaned_df.nunique()
```

```python
#Handling missing data
round((cleaned_df.isnull().sum()/cleaned_df.shape[0])*100,2)
```

```python
cleaned_df.info()
cleaned_df['Date'] = cleaned_df['Date'].dt.date
```

```python
# clustering process below does not take datetime values
cleaned_df=cleaned_df.drop(columns=["Date"], axis=1)
```

```python
# removing outliers if present. Clustering doesn't work as all values are -1. Stock prices cannot be grouped in clusters as they are very fluctuating and volatile

from sklearn.cluster import DBSCAN

pd.options.mode.chained_assignment = None  # default='warn'
cleaned_df['Cluster']= 0

dbscan = DBSCAN(eps=5, min_samples=3)
cleaned_df['Cluster'] = dbscan.fit_predict(cleaned_df)

# Extracting outliers (noise points)
outliers = cleaned_df[cleaned_df['Cluster'] == -1]
outliers
#cleaned_df[:20]

# Removing outliers from the original DataFrame
#cleaned_df = df[df['Cluster'] != -1].drop('Cluster', axis=1)
```

```python
# removing using linear interpolation. Linear interpolation is sensitive to outliers because it calculates values based on a straight line between adjacent data points.

threshold = 10
outliers = cleaned_df['Open'] > threshold

# Applying linear interpolation to handle outliers
cleaned_df['Cleaned_Value'] = np.where(outliers, np.nan, cleaned_df['Open'])
cleaned_df['Cleaned_Value'].interpolate(method='linear', inplace=True)

# Displaying the result
print("Original DataFrame:")
print(cleaned_df)
```

```python
cleaned_df=cleaned_df.drop(columns=["Cluster", "Cleaned_Value"], axis=1)
```

```python
#scaling the data

from sklearn.preprocessing import MinMaxScaler

# initialising the MinMaxScaler
scaler = MinMaxScaler(feature_range=(1, 1000))

# Numerical columns
num_col_ = [col for col in cleaned_df.columns if cleaned_df[col].dtype != 'object']
x1 = cleaned_df
# learning the statistical parameters for each of the data and transforming
x1[num_col_] = scaler.fit_transform(x1[num_col_])
pd.set_option('display.precision', 8)
x1.head()
```
